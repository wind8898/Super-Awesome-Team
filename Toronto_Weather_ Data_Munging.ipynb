```python
import pandas as pd
```

```python
file = 'Toronto_Weather_2012Nov-2019Nov.csv'
```

```python
df = pd.read_csv(file)
df.info(2)
df.head(1)
```

```python
# Re-format the "Date" Column 
df_iso = df["dt_iso"]
df_iso = df_iso.str.split(expand=True)
df_iso = df_iso.rename(columns={0:"Date", 1:"Time"})
df_iso.head()
```

```python
# Create a customized Dataframe
df_new = df_iso.join(df)
df_new = df_new[["Date", "Time", "temp", "weather_main", "rain_1h", "rain_3h", "rain_24h", "rain_today"]]
df_new.info()
```

```python
# Select 2017 data only
df_new['Date'] = pd.to_datetime(df_new['Date'])
condition = (df_new['Date'] >= '2017-01-01') & (df_new['Date'] <= '2017-12-31')
Toronto_Weather_2017 = df_new.loc[condition]
Toronto_Weather_2017.head()
```

```python
# Export 2017 Toronto Weather CSV
Toronto_Weather_2017.to_csv("Toronto_Weather_2017.csv")
```
