```python
import csv
import pandas as pd
import matplotlib.pyplot as plt

q1_file = "Bikeshare Ridership (2017 Q1).csv"
q2_file = "Bikeshare Ridership (2017 Q2).csv"
q3_file = "Bikeshare Ridership (2017 Q3).csv"
q4_file = "Bikeshare Ridership (2017 Q4).csv"

q1_df = pd.read_csv(q1_file)
q2_df = pd.read_csv(q2_file)
q3_df = pd.read_csv(q3_file)
q4_df = pd.read_csv(q4_file)

total_2017_df = pd.concat([q1_df, q2_df, q3_df, q4_df], ignore_index = True, sort=False)
total_2017_df.head()
```

```python
#convert the trip start date time to a date
total_2017_df["trip_date"] = [time.date() for time in total_2017_df["trip_start_time"].astype('datetime64[ns]')]


total_2017_gr = total_2017_df.groupby(["trip_date", "user_type"])

trip_summary = pd.DataFrame({"total_trips": total_2017_gr["trip_id"].count(),
                    "total_trip_duration_hours":total_2017_gr["trip_duration_seconds"].sum()/60/60,
                    "average_trip_duration_minutes": total_2017_gr["trip_duration_seconds"].mean()/60})
```

```python

trip_summary.to_csv("bike_data_summary.csv", index=True)
```
