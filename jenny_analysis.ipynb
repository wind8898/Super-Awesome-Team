```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Take in all of our bikeshare data and read it into pandas
bikeshare_q1 = '2017_Data/Bikeshare Ridership (2017 Q1).csv'
bikeshare_q2 = '2017_Data/Bikeshare Ridership (2017 Q2).csv'
bikeshare_q3 = '2017_Data/Bikeshare Ridership (2017 Q3).csv'
bikeshare_q4 = '2017_Data/Bikeshare Ridership (2017 Q4).csv'

bikeshare_q1_df = pd.read_csv(bikeshare_q1)
bikeshare_q2_df = pd.read_csv(bikeshare_q2)
bikeshare_q3_df = pd.read_csv(bikeshare_q3)
bikeshare_q4_df = pd.read_csv(bikeshare_q4)
```

```python
# Merge the four datasets
bikeshare_2017_df = pd.concat([bikeshare_q1_df, bikeshare_q2_df, bikeshare_q3_df, bikeshare_q4_df], sort=True)
bikeshare_2017_df.head(10)
```

```python
# Find all the unique station names from 'from_station_name' column
from_station_name_df = bikeshare_2017_df['from_station_name'].unique().tolist()
bikeshare_2017_df['from_station_name'].unique().tolist()
```

```python
# Find all the unique station names from 'to_station_name' column
bikeshare_2017_df['to_station_name'].unique().tolist()
to_station_name_df = bikeshare_2017_df['to_station_name'].unique().tolist()
```

```python
stn_name_df = [from_station_name_df + to_station_name_df]
stn_name_df = pd.DataFrame(stn_name_df)
station_name_df = stn_name_df.T
#station_name_df = station_name_df.split('/',expand=True)
#station_name_df = station_name_df.str.split('-', n=1).str[-1]
station_name_df.dropna()
```
